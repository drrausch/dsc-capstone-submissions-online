### Google Trend Data | Popular Design Styles

Finding design aesthetics that have more recently reached peak popularity. Working under the assumption that aesthetics that have recently reached peak popularity have less understanding than those that have been around for much longer. 

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
# for name, filepath in directory.items():
#         name = pd.read_csv(filepath).iloc[1:]
```

```python
art_deco = pd.read_csv('Google_Trends/Art_Deco.csv').iloc[1:]
boho = pd.read_csv('Google_Trends/Boho-Chic.csv').iloc[1:]
farmhouse = pd.read_csv('Google_Trends/Farmhouse.csv').iloc[1:]
hygge = pd.read_csv('Google_Trends/Hygge.csv').iloc[1:]
mcm = pd.read_csv('Google_Trends/MCM.csv').iloc[1:]
minimal = pd.read_csv('Google_Trends/Minimalism.csv').iloc[1:]
modern = pd.read_csv('Google_Trends/Modern.csv').iloc[1:]
rustic = pd.read_csv('Google_Trends/Rustic.csv').iloc[1:]
scandi = pd.read_csv('Google_Trends/Scandinavian.csv').iloc[1:]
wabi_sabi = pd.read_csv('Google_Trends/Wabi-Sabi.csv').iloc[1:]
```

```python
art_deco['Category: All categories'] = art_deco['Category: All categories'].astype('int64')
boho['Category: All categories'] = boho['Category: All categories'].astype('int64')
farmhouse['Category: All categories'] = farmhouse['Category: All categories'].astype('int64')
mcm['Category: All categories'] = mcm['Category: All categories'].astype('int64')
minimal['Category: All categories'] = minimal['Category: All categories'].astype('int64')
modern['Category: All categories'] = modern['Category: All categories'].astype('int64')
rustic['Category: All categories'] = rustic['Category: All categories'].astype('int64')
scandi['Category: All categories'] = scandi['Category: All categories'].astype('int64')
wabi_sabi['Category: All categories'] = wabi_sabi['Category: All categories'].astype('int64')
```

```python
hygge.info()
```

```python
hygge['Category: All categories'].replace(to_replace = '<1', value = '1', inplace = True)
```

```python
hygge.head()
```

```python
hygge['Category: All categories'] = hygge['Category: All categories'].astype('int64')
```

```python
styles = [art_deco, boho, farmhouse, hygge, mcm, minimal, modern, rustic, scandi, wabi_sabi]

styles_dict = {'art_deco': art_deco, 
            'boho': boho,
            'farmhouse': farmhouse,
            'hygge': hygge,
            'mcm': mcm,
            'minimal': minimal,
            'modern': modern,
            'rustic': rustic,
            'scandi': scandi,
            'wabi_sabi': wabi_sabi}

for name,trend in styles_dict.items(): 
    line = trend.plot(title = name)
```
